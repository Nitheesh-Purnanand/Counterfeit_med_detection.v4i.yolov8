```python
import pandas as pd
import numpy as np

df = pd.read_csv('medical_data.csv')
df
```

```python
df["GlucoseGroup"] = df["FastingGlucose"].apply(lambda x:"Normal" if x<100 else "High")
df["BMIGroup"] = df["BMI"].apply(lambda x:"Normal" if x<25 else "Overweight")
```

```python
df
```

```python
def calculate_entropy(series):
    probs = series.value_counts(normalize=True)
    return -np.sum(probs * np.log2(probs))
```

```python
def calculate_conditional_entropy(df, feature, target):
    values = df[feature].unique()
    total_entropy = 0
    total_samples = len(df)
    
    for val in values:
        subset = df[df[feature] == val][target]
        weight = len(subset) / total_samples
        total_entropy += weight * calculate_entropy(subset)
    
    return total_entropy


print(calculate_entropy(df['HeartDisease']))
print(calculate_conditional_entropy(df, 'GlucoseGroup', 'HeartDisease'))

# features = [ 'GlucoseGroup','BMIGroup', 'Smoking']
# results = []

# for col in features:
#     h_y_x = calculate_conditional_entropy(df, col, 'HeartDisease')
#     mi = h_y - h_y_x
#     results.append({'Feature': col, 'H(Y|X)': h_y_x, 'MI': mi})

# results_df = pd.DataFrame(results)
# print("\nMutual Information Results : ")
# print(results_df)
```
