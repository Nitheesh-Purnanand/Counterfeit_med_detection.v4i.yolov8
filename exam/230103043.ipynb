```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Assessment-III.csv")
```

```python
def fill(df, column):
    skewness = df[column].skew()
    if skewness < -0.5 or skewness > 0.5:
        df[column].fillna(df[column].median(), inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)
        print(f"Filled missing values in {column} with mean.")

df["annual_income_usd"] = pd.to_numeric(df["annual_income_usd"])
df["age"] = pd.to_numeric(df["age"])
df["spending_score"] = pd.to_numeric(df["spending_score"])
df["website_visits_per_month"] = pd.to_numeric(df["website_visits_per_month"])
df["avg_purchase_value_usd"] = pd.to_numeric(df["avg_purchase_value_usd"])
df["days_since_last_purchase"] = pd.to_numeric(df["days_since_last_purchase"])



fill(df, "annual_income_usd")
fill(df, "age")
fill(df, "spending_score")
fill(df, "website_visits_per_month")
fill(df, "avg_purchase_value_usd")
fill(df, "days_since_last_purchase")
```
